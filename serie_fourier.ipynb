# Décomposition en série de Fourier

Toute fonction périodique $f(t)$ de période $T_0 = \frac{1}{f_0}$ peut se décomposer sous la forme d'une somme de sinus et de cosinus de fréquences mulitplies de $f_0$ :
\begin{equation*}
f(t) = a_0 + \sum_{n=1}^{\infty} a_i cos(2\pi n f_0 t) + \sum_{n=1}^{\infty} b_i sin(2\pi n f_0 t) 
\end{equation*}

Ou, sachant qu'un cosinus est un sinus décalé :
\begin{equation*}
f(t) =  \sum_{n=0}^{\infty} a_i sin(2\pi n f_0 t + \phi_i) 
\end{equation*}

avec :
- $n f_0$ : la fréquence des sinus
- $a_i$ : l'amplitude de chaque sinus
- $\phi_i$ : la phase de chaque sinus


```python
import numpy as np 
import matplotlib.pyplot as plt
from plot_result import f1
%matplotlib inline

# --- MODIFIER ICI l'AMPLITUDE ET LA PHASE DE CHAQUE SINUS ----
A = (1,1,0) # Amplitude en unité arbitraire
PH = (0,0,0) # phase en degre
# -------------------------------------------------------------

f0 = .2
t = np.arange(0,10,.01)
f = np.zeros_like(t)
for a,ph,i in zip(A,PH,(1,2,3)):
    floc = a*np.sin(2* np.pi * i * f0 * t+ph*np.pi/180.)
    f = f + floc
    plt.plot(t,floc,'b')
```
